```
import torch
from transformers import AutoModel, AutoTokenizer, BertTokenizer
import pandas as pd
torch.set_grad_enabled(False) 
```

```
from transformers import DistilBertModel

bert_distil = DistilBertModel.from_pretrained('distilbert-base-cased')
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased")
```

```
conv_logs = pd.read_csv("sample_clustering.csv")
```

```
output_embeddings = []
count = 0
for row in conv_logs.utterance:
  count +=1
  tokens = tokenizer(row, return_tensors="pt")
  embedding = bert_distil(tokens['input_ids'])
  print(str(count) + ": " + str(embedding))
  print("Length Input: ", len(embedding[0][0]))
  output_embeddings.append(embedding)
```

```
print(output_embeddings[0][0].tolist())
```

```
embedding_df = pd.DataFrame(columns= ["UtteranceEmbeddings"])
for x in range(len(output_embeddings)):
  embedding_df.at[x, 'UtteranceEmbeddings'] = output_embeddings[x][0].tolist()
  print(embedding_df.loc[x, 'UtteranceEmbeddings'])
  # embedding_df.set_value('UtteranceEmbeddings', 'x', 10)
```
